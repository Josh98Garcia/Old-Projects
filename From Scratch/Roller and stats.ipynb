```
import random
```

```
def rolld6():
  return random.choices(range(1, 7), weights=(.05, .1, .175, .125, .25, .2))
```

```
def roll():
  return rolld6() + rolld6()
```

```
def rolladd():
  r = roll()
  return r[0] + r[1]
```

```
def fair():
  r = random.choices(range(1, 7)) + random.choices(range(1, 7))
  return r[0] + r[1]
```

```
 
def test():
  result = [0, 0, 0]
  resultfair = [0, 0, 0]
  l1 = list(range(1000))
  l2 = list(range(1000))
  summed = 0
  sumfair = 0
  for x in list(range(1000)):
    l1[x] = rolladd()
    l2[x] = fair()
 
  for x in l2:
    sumfair += x
    if x <= 6:
      resultfair[0] += 1
    elif x <= 9:
      resultfair[1] += 1
    else:
      resultfair[2] += 1
  
  for x in l1:
    summed += x
    if x <= 6:
      result[0] += 1
    elif x <= 9:
      result[1] += 1
    else:
      result[2] += 1
    
  #print(summed, result)
  #print(sumfair, resultfair)
  return summed, result[0], result[1], result[2], sumfair, resultfair[0], resultfair[1], resultfair[2]
```

```
 def t2():
  results = [0, [0, 0, 0], 0, [0, 0, 0]]
  summed, result[0], result[1], result[2], sumfair, resultfair[0], resultfair[1], resultfair[2] = test()
  results[0] += summed
  for x in list(range(3)):
    results[1][x] += result[x]
  results[2] += sumfair
  for x in list(range(3)):
    results[3][x] += resultfair[x]
  return results
```
